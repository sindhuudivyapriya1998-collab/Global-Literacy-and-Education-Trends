# Global Literacy & Education Trends

```python
import pandas as pd
df_adult =pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_adult
```

```python
df_adult.to_csv("literacy_rate_adults.csv")
```

```python
df_illiteracy =pd.read_csv("https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_illiteracy
```

```python
df_schooling =pd.read_csv("https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_schooling
```

```python
df_youth = pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_youth
```

```python
df_youth.to_csv=("youth.csv")
```

```python
df_gdp = pd.read_csv("https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_gdp
```

```python
df_gdp.to_csv("gdp.csv")
```

```python
import pandas as pd
df_literacy = pd.merge(df_adult,df_youth,on=["entity","code","year"],how="outer")
df_literacy
```

```python
df_gdp_schooling=pd.merge(df_gdp,df_schooling,on=["entity","code","year"],how="outer")
df_gdp_schooling
```

#Data Cleaning

```python
df_literacy=df_literacy.dropna()
```

```python
df_literacy.rename(columns={"entity":"country","adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99":"adult_literacy_rate","youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m":"youth_literacy_male","youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f":"youth_literacy_female","owid_region":"region"},inplace=True)
df_literacy
```

```python
df_literacy['Avg_youth_literacy'] =(df_literacy['youth_literacy_male'] + df_literacy['youth_literacy_female'])/2
df_literacy
```

```python
df_literacy = df_literacy.sort_values(['country','year'])
df_literacy
```

```python
df_literacy['youth_growth_rate'] = (df_literacy.groupby('country')['Avg_youth_literacy'].pct_change()*100)
df_literacy
```

```python
df_literacy.dropna()
```

```python
df_illiteracy.dropna()
```

```python
df_illiteracy.rename(columns={"entity":"country"},inplace=True)
df_illiteracy
```

```python
df_gdp_schooling.rename(columns={"entity":"country","ny_gdp_pcap_pp_kd":"gdp_per_capita","mf_youth_and_adults__15_64_years__average_years_of_education":"Avg_years_schooling","owid_region_x":"region_gdp","owid_region_y":"region_schooling"},inplace=True)
df_gdp_schooling
```

```python
df_gdp_schooling=df_gdp_schooling.dropna()
```

```python
df_gdp_schooling=df_gdp_schooling.reset_index(drop=True)
df_gdp_schooling
```

```python
df_literacy = df_literacy[(df_literacy['year'] >= 1990)&(df_literacy['year'] <= 2025)]
df_literacy
```

```python
df_literacy['literacy_gender_gap'] = (df_literacy['youth_literacy_male'] - df_literacy['youth_literacy_female']).abs()
df_literacy
```

```python
df_illiteracy=df_illiteracy[(df_illiteracy['year'] >= 1990)&(df_illiteracy['year'] <= 2025)].reset_index(drop=True)
df_illiteracy
```

```python
df_gdp_schooling=df_gdp_schooling[(df_gdp_schooling['year'] >= 1990)&(df_gdp_schooling['year'] <= 2025)]
df_gdp_schooling
```

```python
df_gdp_schooling["gdp_per_schooling_year"] = df_gdp_schooling["gdp_per_capita"]/df_gdp_schooling["Avg_years_schooling"]
df_gdp_schooling
```

```python
df_gdp_schooling["education_index"]=(df_gdp_schooling["literacy_rate"]+(df_gdp_schooling["Avg_years_schooling"]*10))/2
df_gdp_schooling
```

```python
df_literacy.to_csv("literacy.csv")
```

```python
df_illiteracy.to_csv("illiteracy.csv")
```

```python
df_gdp_schooling.to_csv("Gdp.csv")
```
